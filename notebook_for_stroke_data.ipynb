```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns

from sklearn.model_selection import cross_val_score, RepeatedStratifiedKFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
```

```python
# Reading csv file containing stroke data

stroke_data = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
# drop ID column

stroke_data = stroke_data.drop("id", axis=1)
stroke_data.tail()
```

```python
# Percentage of missing data in column

missing_data = (stroke_data["bmi"].isnull().sum()/len(stroke_data["bmi"]))*100
print(f"Missing BMI data: {missing_data:.2f}%")
```

```python
# Check distribution of bmi to choose appropriate imputation method for replacing missing values

sns.histplot(data=stroke_data["bmi"])
plt.axvline(stroke_data.bmi.mean(), color='r')
plt.axvline(stroke_data.bmi.median(), color='y')
print(f"Mean: {stroke_data.bmi.mean()}, Median: {stroke_data.bmi.median()}")
plt.title("BMI Distribution showing median value in yellow and mean value in red")
plt.show()
```

```python
# Replace NaN values with column average using imputer with the median method due to slight skew in distribution

stroke_data["bmi"] = stroke_data["bmi"].fillna(value=stroke_data.bmi.median())
stroke_data["bmi"].head()
```

```python
stroke_only = categorical_data.loc[categorical_data['stroke'] == 'Yes']
stroke_only.shape
```

```python
# Counting values within categorical data only considering those that have suffered a stroke

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(14,8))
ax = np.ravel(ax)


for i, col in enumerate(stroke_only):
    plt.sca(ax[i])
    sns.countplot(x=col, data=stroke_only)
    plt.xticks(rotation=90)
    
    

plt.tight_layout()
```

Although those which have suffered a stroke don't show an overwhelming bias in terms of hypertension and heart disease, the difference betweent he Yes/No peaks compared with those in the entire dataset are relatively smaller meaning these columns do play a larger role.

We can solve this visually by dividing through the entire dataset for each point, i.e., stroke only hypertension data / whole hypertension data.

This needs to be performed in the case due to the lacking number of people that have suffered a stroke within the dataset.

```python
# Function to return dictionary of fixed values for each of the categorical data

def weighted_vals(col_name):
    weighted_predictors = {}
    for i, j in (stroke_only[col_name].value_counts() / categorical_data[col_name].value_counts()).iteritems():
        weighted_predictors[i] = j
    return weighted_predictors
```

```python
# Counting values within categorical data after considering entire dataset

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(14,8))
ax = np.ravel(ax)


for i, col in enumerate(categorical_data):
    val_dict = weighted_vals(col)
    plt.sca(ax[i])
    sns.barplot(x=list(val_dict.keys()), y=list(val_dict.values()))
    plt.title(col)
    plt.xticks(rotation=90)   

plt.tight_layout()
```

The above barplots give a clearer picture of good predictors since it takes into account biasses due to certain columns lacking variety.

```python
# Converting ever_married column to 0's and 1's

stroke_data_original["ever_married"].replace(("Yes", "No"), (1, 0), inplace=True) 

stroke_data_original.head()
```

```python
# Seperating predictors

X = stroke_data_original.drop("stroke", axis=1)
y = stroke_data_original["stroke"]
```

```python
# Preprocessing of numerical data

numerical_train_data = X.select_dtypes(include=["float64"]).columns
categorical_train_data = X.select_dtypes(include=["object", "int64"]).columns

# Preprocessing of numerical data
numerical_transformer = StandardScaler()

# Preprocessing of categorical data
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Bundel preprocessing for numerical and categorical 
preprocessor = ColumnTransformer(transformers = [
    ('num', numerical_transformer, numerical_train_data),
    ('cat', categorical_transformer, categorical_train_data),

])
```

```python
# Initialising ML models

log_reg = LogisticRegression()
k_neighbours = KNeighborsClassifier()
dec_tree = DecisionTreeClassifier()
random_forest = RandomForestClassifier()
svc = SVC()
xgb = XGBClassifier(use_label_encoder=False)

model_list = [log_reg, k_neighbours, dec_tree, random_forest, svc, xgb]
model_titles = ["Logistic Regression", "K-nearest Neighbours", "Decision Tree", "Random Forest", "SVC", "XGBoost"]
```

```python
i = 0
model_scores_dict = {}

for models in model_list:
    my_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('model', models)
    ])
    cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=42)
    model_score = cross_val_score(my_pipeline, X, y,
                              cv=cv,
                              scoring='accuracy')
    model_scores_dict[model_titles[i]] = model_score.tolist()
    i += 1

print(model_scores_dict)
```

```python
plt.plot(model_scores_dict.keys(), [np.mean(i) * 100 for i in model_scores_dict.values()], 'ob--')
plt.xticks(rotation=45)
plt.ylabel("Accuracy (%)")
plt.show()
```
