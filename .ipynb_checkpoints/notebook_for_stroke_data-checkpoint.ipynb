```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LogisticRegression
```

```python
# Reading csv file containing stroke data

stroke_data = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
stroke_data.head()
```

```python
# drop ID column

stroke_data = stroke_data.drop("id", axis=1)
stroke_data.tail()
```

```python
# Check for missing values in dataset

stroke_data.isnull().sum()
```

```python
# Percentage of missing data in column

missing_data = (stroke_data["bmi"].isnull().sum()/len(stroke_data["bmi"]))*100
print(f"Missing BMI data: {missing_data:.2f}%")
```

```python
# Check distribution of bmi to choose appropriate imputation method for replacing missing values

sns.histplot(data=stroke_data["bmi"])
plt.axvline(stroke_data.bmi.mean(), color='r')
plt.axvline(stroke_data.bmi.median(), color='y')
print(f"Mean: {stroke_data.bmi.mean()}, Median: {stroke_data.bmi.median()}")
plt.title("BMI Distribution showing median value in yellow and mean value in red")
plt.show()
```

```python
# Replace NaN values with column average using imputer with the median method due to slight skew in distribution

stroke_data["bmi"] = stroke_data["bmi"].fillna(value=stroke_data.bmi.median())
stroke_data["bmi"].head()
```

```python
stroke_data.dtypes
```

```python
# Converting 0 -> No and 1 -> Yes for categorical columns

for column in stroke_data:
    if stroke_data[column].dtype == "int64":
        stroke_data[column].replace((1, 0), ("Yes", "No"), inplace=True) 

stroke_data.head()
```

```python
stroke_data.dtypes
```

```python
# float64 type data

numerical_data = stroke_data.select_dtypes(include=["float64"])
categorical_data = stroke_data.select_dtypes(include=["object"])

numerical_data.head()
```

```python
categorical_data.head()
```

```python
# Heatmap plot to check correlations in numerical data

plt.figure(figsize=(6, 5))
sns.heatmap(data=numerical_data.corr(), annot=True)
plt.show()
```

```python
# Pairwise scatter plots to examine linear relationships

sns.pairplot(data=numerical_data, kind="reg", diag_kind="kde",
            plot_kws={'line_kws':{'color':'lime'}, 'scatter_kws': {'alpha': 0.1, "marker": "+"}})
```

```python
# Adding hues to our pairplot - Heart Disease

sns.pairplot(data=stroke_data, kind="scatter", diag_kind="kde", hue="heart_disease",
            plot_kws={'alpha': 1, "marker": "o"})
```

```python
# Adding hues to our pairplot - Married

sns.pairplot(data=stroke_data, kind="scatter", diag_kind="kde", hue="ever_married",
            plot_kws={'alpha': 1, "marker": "o"})
```

```python
# Adding hues to our pairplot - Smoking status

sns.pairplot(data=stroke_data, kind="scatter", diag_kind="kde", hue="smoking_status",
            plot_kws={'alpha': 1, "marker": "o"})
```

```python
# Adding hues to our pairplot - Stroke (predictor target)

sns.pairplot(data=stroke_data, kind="scatter", diag_kind="kde", hue="stroke",
            plot_kws={'alpha': 1, "marker": "o"})
```

Age plays a key role in predicting a stroke.

```python
# Counting values within categorical data

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(14,8))
ax = np.ravel(ax)


for i, col in enumerate(categorical_data):
    plt.sca(ax[i])
    sns.countplot(x=col, data=categorical_data)
    plt.xticks(rotation=90)
    

plt.tight_layout()
```

```python
stroke_only = categorical_data.loc[categorical_data['stroke'] == 'Yes']
stroke_only.shape
```

```python
# Counting values within categorical data only considering those that have suffered a stroke

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(14,8))
ax = np.ravel(ax)


for i, col in enumerate(stroke_only):
    plt.sca(ax[i])
    sns.countplot(x=col, data=stroke_only)
    plt.xticks(rotation=90)
    
    

plt.tight_layout()
```

Although those which have suffered a stroke don't show an overwhelming bias in terms of hypertension and heart disease, the difference betweent he Yes/No peaks compared with those in the entire dataset are relatively smaller meaning these columns do play a larger role.

```python
hypertension_yes = categorical_data.loc[categorical_data['hypertension'] == 'Yes']
hypertension_no = categorical_data.loc[categorical_data['hypertension'] == 'No']

print(f"Percentage of people who have hypertension that also suffered a stroke: {(stroke_only['hypertension'].shape[0] / hypertension_yes['hypertension'].shape[0]) * 100}%")
print(f"Percentage of people who don't have hypertension that suffered a stroke: {(stroke_only['hypertension'].shape[0] / hypertension_no['hypertension'].shape[0]) * 100}%")
```

```python
heart_disease_yes = categorical_data.loc[categorical_data['heart_disease'] == 'Yes']
heart_disease_no = categorical_data.loc[categorical_data['heart_disease'] == 'No']

print(f"Percentage of people who have heart disease that also suffered a stroke: {(stroke_only['heart_disease'].shape[0] / heart_disease_yes['heart_disease'].shape[0]) * 100}%")
print(f"Percentage of people who don't have heart disease that suffered a stroke: {(stroke_only['heart_disease'].shape[0] / heart_disease_no['heart_disease'].shape[0]) * 100}%")
```
