```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import mean_absolute_error
from sklearn.impute import SimpleImputer
```

```python
# Reading csv file containing stroke data

stroke_data = pd.read_csv("healthcare-dataset-stroke-data.csv")
```

```python
stroke_data.head()
```

```python
# drop ID column

stroke_data = stroke_data.drop("id", axis=1)
stroke_data.head()
```

```python
# Check for missing values in dataset

stroke_data.isnull().sum()
```

```python
# Percentage of missing data in column

missing_data = (stroke_data["bmi"].isnull().sum()/len(stroke_data["bmi"]))*100
print(f"Missing data: {missing_data}%")
```

```python
# Check distribution of bmi to choose appropriate imputation method for replacing missing values

sns.histplot(data=stroke_data["bmi"])
print(f"Mean: {stroke_data.bmi.mean()}, Median: {stroke_data.bmi.median()}")
```

```python
# Replace NaN values with column average using imputer with the median method due to slight skew in distribution

stroke_data["bmi"] = stroke_data["bmi"].fillna(value=stroke_data.bmi.median())
stroke_data["bmi"].head()
```

```python
# Check data is clean

stroke_data.head()
```

```python
stroke_data.dtypes
```

```python
# float64 type data

numerical_data = stroke_data.select_dtypes(include=["float64", "int64"])
numerical_data.head()
```

```python
# Heatmap plot to check correlations in numerical data

#plt.figure(figsize=(9, 9))

sns.heatmap(data=numerical_data.corr(), annot=True)
plt.show()
```

```python
# Highest correlations with age
# Checking for correlation with gender
# bmi vs age

fig, ax = plt.subplots(1, 2, figsize=(12,6))
fig.tight_layout()
fig.suptitle("BMI and average glucose plotted against age")
fig.subplots_adjust(top=0.9)

sns.scatterplot(x=stroke_data["age"], y=stroke_data["bmi"], hue=stroke_data["gender"], ax=ax[0])
ax[0].set_title("BMI")
sns.scatterplot(x=stroke_data["age"], y=stroke_data["avg_glucose_level"], hue=stroke_data["gender"], ax=ax[1])
ax[1].set_title("Glucose level")

```

```python
# bmi vs age - box plots

fig, ax = plt.subplots(1, 2, figsize=(12,6))
fig.tight_layout()
fig.suptitle("BMI and average glucose plotted box plots with seperate genders")
fig.subplots_adjust(top=0.9)

sns.boxenplot(x="gender", y="bmi", data=stroke_data, ax=ax[0])
ax[0].set_title("BMI")
sns.boxenplot(x="gender", y="avg_glucose_level", data=stroke_data, ax=ax[1])
ax[1].set_title("Glucose level")
```
